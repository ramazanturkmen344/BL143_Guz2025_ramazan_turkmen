```python
print("=== GÜNLÜK PLANLAYICI ===")

while True:
    while True:
        try:
            uyku = int(input("Günde kaç saat uyuyorsunuz? "))
            if uyku < 0:
               print("Lütfen pozitif bir sayi girin\n")
               continue
        except ValueError:
            print("Lütfen tam sayı girin\n")
            continue
        break
    while True:
        try:
            ders = int(input("Günde kaç saat ders çalışıyorsunuz? "))
            if ders < 0:
               print("Pozitif bir sayı girin\n")
               continue
        except ValueError:
            print("Lütfen tam sayı girin\n")
            continue
        break
    while True:
        try:
            serbest = int(input("Günde kaç saat kendinize zaman ayırıyorsunuz? "))
            if serbest < 0:
                print("Pozitif bir sayı girin.\n")
                continue
        except ValueError:
            print("Lütfen tam sayı girin\n")
            continue
        break
    if uyku + ders + serbest > 24:
        print("Lütfen saatlerin toplamı 24 saati geçmemek üzere tekrar girin.\n")
        continue
    break
if uyku < 6:
    uykudurum = "Yetersiz uyku"
elif uyku >= 6 and uyku <= 8:
    uykudurum = "Normal uyku"
else:
    uykudurum = "Fazla uyku"
print(f"Uyku durumu: {uykudurum}\n")


if  ders < 6:
    dersdurum = "Normal çalışma günü\n"
elif ders >= 6:
    dersdurum = "Yoğun çalışma günü\n"
print(f"Ders durumunuz: {dersdurum}")

if uyku < 6 and ders >= 6:
    print ("Verimsiz bir gün geçirmişsiniz\n")
else:
    print("Dengeli bir gün geçirmişsiniz\n")




 # Değişken: Verilen bilgiyi atadığımız isim için hafıza da tutar.
 # Koşul: Yapılacak iş için duruma göre verdiğimiz yol ayrımı.
 # Karşılaştırma Operatörü: Değerleri birbiriyle kıyaslayıp sonucu veren sembollerdir.
 # Girdi / Çıktı: Progamcı ile kullanıcı arasındaki iletişim.
 # Hata Kontrölü: Beklenmeyen bir durum programın çökmesini engelleyen güvenlik sistemi.
```
